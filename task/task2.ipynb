```python
#  Write a Python program to do the following: 
# • Accept lengths L1 and L2 for two lists from the user. 
# • Create List A of length L1 and ensure that there are no repeating elements 
# • Create List B of length L2 and ensure that there are no repeating elements 
# • Perform the following: o Create list C that contains elements common to lists A and B. 
# a. Create list D that contains elements unique to lists A and B. 
# b. Print list C 
# c. Print list D 
# d. Print C sorted in ascending order.
# e. Print D sorted in descending order 

L1=int(input("Enter L1:"))
L2=int(input("Enter L2:"))
A,B,C,D=[],[],[],[]
print(f"L1={L1} L2={L2}")
for x in [L1,L2]:
    i=0
    while i<x:
        a=int(input(f"enter {i+1} element:"))
        if x==L1:
            if a not in A:
                A.append(a)
                i+=1
            else:
                print(f"{a} is already in the list enter another number")
        else:
            if a not in B:
                B.append(a)
                i+=1
            else:
                print(f"{a} is already in the list enter another number")
                
print(f"A={A}")
print(f"B={B}")
A=set(A)
B=set(B)
C=A&B
D=A^B
print(f"Intersection C={C}")
print(f"Symmetric Difference D={D}")

print(sorted(C))
print(sorted(D,reverse=True))
```

```python
# Write a Python function that takes a list of numbers and determines if all the numbers are unique. If yes, return 
# TRUE, else return FALSE

def unique(li):
    for i in range(len(li)):
        if i in li[i+1:]:
            return True
    return False

li=list(map(int,input().split()))
print(li)
unique(li)

```

```python
#  Create a Python class “BankAccount” with withdraw() and deposit() methods. Then create 2 users “sam” and 
# “rio”. Users will deposit and withdraw some amount and program should display their current balance.
#      Input          Output 
# sam.deposit(100)      100 
# rio.deopsit(200)      200 
# sam.withdraw(20)       80 
# rio.withdraw(30)       170 

class BankAccount:
    def __init__(self,balance=0):
        self.balance=balance
    
    def withdraw(self,a):
        if self.balance<a:
            print("insufficient balance")
        else:
            self.balance-=a
            print(self.balance)
    def deposit(self,a):
        self.balance+=a
        print(self.balance)
    
sam=BankAccount()
rio=BankAccount()

sam.deposit(100)
rio.deposit(200)
sam.withdraw(20)
rio.withdraw(30)

```

```python
# Implement Insertion Sort in Python
# Print the following:
# • Number of iterations.
# • Number of comparisons
# • Number of swaps

l=[23,-5,3,2,1,-3,-3,7,2,2]
swaps,iterations,comparisons=0,0,0
def insertion_sort(arr):
  global swaps,iterations,comparisons
  n=len(arr)
  for i in range(1,n):
    iterations+=1
    for j in range(i,0,-1):
      comparisons+=1
      if arr[j-1] >arr[j]:
        arr[j-1],arr[j]=arr[j],arr[j-1]
        swaps+=1
      else:
        break

insertion_sort(l)
print(l)
print(f"Number of iterations={iterations}")
print(f"Number of comparisons={comparisons}")
print(f"Number of swaps={swaps}")
```
